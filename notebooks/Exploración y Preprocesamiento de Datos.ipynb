# Handle Missing Values
Identify missing values and apply appropriate strategies to handle them (imputation or removal).

```python
# Check for missing values in the dataset
missing_values = dataset.isnull().sum()

# Display the count of missing values for each column
print("\nMissing Values:")
print(missing_values)

# Visualize missing values using a heatmap (optional, requires seaborn)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(dataset.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()

# Handle missing values
# Example: Drop rows with missing values
dataset_dropped = dataset.dropna()

# Example: Fill missing values with the mean (for numerical columns)
dataset_filled = dataset.fillna(dataset.mean())

# Display the shape of the dataset after handling missing values
print("\nDataset Shape After Dropping Missing Values:", dataset_dropped.shape)
print("Dataset Shape After Filling Missing Values:", dataset_filled.shape)
```

# Detect and Handle Outliers
Use statistical methods and visualizations to detect outliers and decide on handling strategies.

```python
import numpy as np

# Detect outliers using the IQR method
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in the dataset
outliers = ((dataset < lower_bound) | (dataset > upper_bound))

# Display the count of outliers for each column
print("\nOutliers Count:")
print(outliers.sum())

# Visualize outliers using boxplots for numerical columns
numerical_columns = dataset.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot((len(numerical_columns) + 2) // 3, 3, i)
    sns.boxplot(y=dataset[column])
    plt.title(f"Boxplot of {column}")
plt.tight_layout()
plt.show()

# Handle outliers by capping them to the lower and upper bounds
dataset_capped = dataset.copy()
for column in numerical_columns:
    dataset_capped[column] = np.where(dataset[column] < lower_bound[column], lower_bound[column],
                                      np.where(dataset[column] > upper_bound[column], upper_bound[column], dataset[column]))

# Display the shape of the dataset after handling outliers
print("\nDataset Shape After Handling Outliers:", dataset_capped.shape)
```

# Normalize and Scale Features
Apply normalization and scaling techniques to prepare features for analysis.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Select numerical columns for normalization and scaling
numerical_columns = dataset_capped.select_dtypes(include=[np.number]).columns

# Normalize features using Min-Max Scaling
scaler_minmax = MinMaxScaler()
dataset_normalized = dataset_capped.copy()
dataset_normalized[numerical_columns] = scaler_minmax.fit_transform(dataset_capped[numerical_columns])

# Display the first few rows of the normalized dataset
print("\nNormalized Dataset (Min-Max Scaling):")
print(dataset_normalized.head())

# Scale features using Standard Scaling (Z-score normalization)
scaler_standard = StandardScaler()
dataset_scaled = dataset_capped.copy()
dataset_scaled[numerical_columns] = scaler_standard.fit_transform(dataset_capped[numerical_columns])

# Display the first few rows of the scaled dataset
print("\nScaled Dataset (Standard Scaling):")
print(dataset_scaled.head())
```

# Encode Categorical Variables
Convert categorical variables into numerical format using encoding techniques.

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Identify categorical columns in the dataset
categorical_columns = dataset.select_dtypes(include=['object', 'category']).columns

# Encode categorical variables using Label Encoding
label_encoded_dataset = dataset.copy()
label_encoders = {}
for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    label_encoded_dataset[column] = label_encoders[column].fit_transform(dataset[column])

# Display the first few rows of the label-encoded dataset
print("\nLabel Encoded Dataset:")
print(label_encoded_dataset.head())

# Encode categorical variables using One-Hot Encoding
one_hot_encoder = OneHotEncoder(sparse=False, drop='first')  # Drop first to avoid multicollinearity
one_hot_encoded_array = one_hot_encoder.fit_transform(dataset[categorical_columns])
one_hot_encoded_columns = one_hot_encoder.get_feature_names_out(categorical_columns)

# Create a DataFrame for the one-hot encoded variables
one_hot_encoded_df = pd.DataFrame(one_hot_encoded_array, columns=one_hot_encoded_columns, index=dataset.index)

# Concatenate the one-hot encoded variables with the original dataset
one_hot_encoded_dataset = pd.concat([dataset.drop(columns=categorical_columns), one_hot_encoded_df], axis=1)

# Display the first few rows of the one-hot encoded dataset
print("\nOne-Hot Encoded Dataset:")
print(one_hot_encoded_dataset.head())
```
